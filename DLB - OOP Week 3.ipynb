### Method

Method itu sebuah action / function di python. Contoh class nya ini bisa ngapain aja.

# illustration.MethodIllustration()

```python
class Manusia:
    def __init__(self, warna_rambut, tinggi, berat): 
        # nanti jangan langsung dikasih input_value parameter, kasih default value dulu di __init__
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat
        
    def cat_rambutnya(self, warna):
        self.warna_rambut = warna
        print(f'Uwow, warna rambutnya jadi {warna}')

    def BMI(self):
        return (self.berat / self.tinggi ** 2) * 100 # rumus BMI berat dibagi tinggi kuadrat
```

### Inisialisasi Object

### Inheritance

Manfaatkan Class / Kodingan orang lain

```python
illustration.InheritanceIllustration()
```

```python
class Mahasiswa(Manusia):
    def __init__(self, warna_rambut, tinggi, berat, universitas):
        super().__init__(warna_rambut, tinggi, berat) # Pakai super() kalau mau access parent class nya
        self.universitas = universitas # Fitur asli dari class Mahasiswa
        
    def pindah_univ(self, universitas_baru):
        self.universitas = universitas_baru
        print(f'Wawawawawwww, pindah univ ke {universitas_baru}')
```

```python
shandy = Mahasiswa('hitam', 178, 84, "UMM")
```

```python
shandy.universitas
```

```python
shandy.pindah_univ("MIT")
```

```python
shandy.universitas
```

# Polymorphism 

Sebuah kelas dapat memiliki method dengan nama yang sama namun dengan susunan parameter yang berbeda

```python
class Indonesia:
    def capital(self):
        print("Jakarta")
    
    def language(self):
        print("Bahasa Indonesia")
    
class UK:
    def capital(self):
        print("London")
    
    def language(self):
        print("English")
    
class Japan:
    def capital(self):
        print("Tokyo")
    
    def language(self):
        print("Japanese")
```

```python
obj_ina = Indonesia()
obj_eng = UK()
obj_jpn = Japan()

for country in (obj_ina, obj_eng, obj_jpn):
    country.capital()
    country.language()
```
