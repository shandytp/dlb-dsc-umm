### Formula Linear Regression 

\begin{equation}
Y = a + bX
\end{equation}

### Deskripsi

\begin{align*}
    Y &= \text{Variable Response atau Variable Akibat (Dependent)}\\
    X &= \text{Variable Predictor atau Variable Penyebab (Independent)}\\
    a &= \text{Konstanta}\\
    b &= \text{Koefisien Regresi (Kemiringan or Slope)}
\end{align*}

### Metode Evaluasi MSE , RMSE (Cost Function)

### Mean Squared Error

\begin{equation*}
MSE = \frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y}_i)^2
\end{equation*}

### Root Mean Squared Error

\begin{equation*}
RMSE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big(\frac{x_i -\hat{x}_i}{N}\Big)^2}}
\end{equation*}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_simple.csv')
df
```

```python
df.plot(x='Tugas', y='NA', style='o')
plt.title('Tugas vs Nilai Akhir')
plt.xlabel('Nilai Tugas')
plt.ylabel('Nilai Akhir')
plt.show()
```

```python
data_train = df[:15]

X_train = data_train['Tugas'].values.reshape(-1, 1)
Y_train = data_train['NA'].values.reshape(-1, 1)
```

```python
data_train
```

```python
data_test = df[15:]

X_test = data_test['Tugas'].values.reshape(-1, 1)
Y_test = data_test['NA'].values.reshape(-1, 1)
```

```python
data_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
df_test = pd.DataFrame({'Ground Truth': Y_test.flatten(), 'Predicted': y_pred.flatten()})
df_test
```

```python
plt.scatter(X_test, Y_test)
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
from sklearn import metrics

print(f'MSE = {metrics.mean_squared_error(Y_test, y_pred)}')
print(f'RMSE = {np.sqrt(metrics.mean_squared_error(Y_test, y_pred))}')
```

### Formula Naive Bayes 

\begin{equation*}
P(A\mid B) = \frac{P(B\mid A)P(A)}{P(B)}
\end{equation*}

### Deskripsi

P(A|B) is Posterior probability: Probability of hypothesis A on the observed event B.

P(B|A) is Likelihood probability: Probability of the evidence given that the probability of a hypothesis is true.

P(A) is Prior Probability: Probability of hypothesis before observing the evidence.

P(B) is Marginal Probability: Probability of Evidence.

```python
df_iris = pd.read_csv('Iris.csv')
df_iris
```

```python
df_iris.drop('Id', axis=1, inplace=True)
```

```python
df_iris
```

```python
df_iris.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_iris.drop('Species', axis=1)
# X = df_iris(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
X
```

```python
y = df_iris[['Species']]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```
